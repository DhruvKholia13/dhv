```python
music_url="https://en.wikipedia.org/wiki/Forbes_list_of_the_world%27s_highest-paid_musicians"
mus_response = requests.get(music_url)
mus_soup = BeautifulSoup(mus_response.text, 'html.parser')
```

```python
print(f"NUMBER OF TABLES IN THE PAGE: {len(mus_soup.find_all('table'))}")
```

```python
table = mus_soup.find_all('table')[1]
print(table)
```

```python
table.find_all('tr')[13].contents
```

```python
table.find_all('tr')[13].text
```

```python
for data in table.find_all('tr')[13].find_all('td'):
    print(data.text)
```

```python
lst = []
for row in table.find_all('tr')[1:-1]:
    s = pd.Series([data.text for data in row.find_all('td')])
    lst.append(s)
```

```python
data = pd.concat(lst, axis=1).T
```
