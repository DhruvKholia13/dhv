```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame(np.random.rand(10,10))
plt.figure(figsize=(8,6))
plt.contour(data.values)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Contour lines with colormap")
plt.colorbar(label="value")
plt.show()
```
